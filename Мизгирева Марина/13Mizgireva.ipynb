```python
#13.Задано множество правильных многоугольников (не более 25),
#для каждого из которых указаны площадь и количество вершин. Реализовать сортировку по площади и периметру,
#а также вычислить суммарную площадь и суммарный периметр.

#многоугольник 1 - треугольник со стороной 3. Площадь = 3,9
l1 = 3 
S1 = 3.896
vertex1 = 3
P1 = l1*vertex1

#многоугольник 2 - квадрат со стороной 5. Площадь = 25
l2 = 5
S2 = 25
vertex2 = 4
P2 = l2*vertex2

#многоугольник 3 - пятиугольник со стороной 2. Площадь = 6,9
l3 = 2
S3 = 6.9
vertex3 = 5
P3 = l3*vertex3

#многоугольник 4 - семиугольник со стороной 15. Площадь = 818.1
l4 = 15
S4 = 818.1
vertex4 = 7
P4 = l4*vertex4

#многоугольник 5 - треугольник со стороной 16. Площадь = 110.9
l5 = 16
S5 = 110.9
vertex5 = 3
P5 = l5*vertex5

list_S = []
list_S += [S1, S2, S3, S4, S5]
sort_S = sorted (list_S)
print('Сортировка площадей правильных многоугольников по возрастанию: ')
print(*sort_S, sep=', ')
#print(list_S)

list_P = []
list_P += [P1, P2, P3, P4, P5]
sort_P = sorted(list_P)
print('Сортировка периметров правильных многоугольников по возрастанию: ')
print(*sort_P, sep=', ')


sum_S = sum(list_S)
print('Суммарная площадь всех правильных многоугольников равна ' + str(sum_S))

sum_P = sum(list_P)
print('Суммарный периметр всех правильных многоугольников равен ' + str(sum_P))


```
