### THE DATA

** Read in the census_data.csv data with pandas**

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
census = pd.read_csv('census_data.csv')
```

```python
census.head()
```

```python
census.isnull().any()
```

```python
census['workclass'].unique(), census['education'].unique(), census['marital_status'].unique(), census['occupation'].unique(), census['relationship'].unique(), census['race'].unique(), census['gender'].unique(), census['native_country'].unique(), census['income_bracket'].unique()
```

** TensorFlow won't be able to understand strings as labels, you'll need to use pandas .apply() method to apply a custom function that converts them to 0s and 1s. This might be hard if you aren't very familiar with pandas, so feel free to take a peek at the solutions for this part.**

** Convert the Label column to 0s and 1s instead of strings.**

```python
census['income_bracket'][census['income_bracket']=='>50K'] = 1
census['income_bracket'][census['income_bracket']=='<=50K'] = 0
```

```python
census.head()
```

### Perform a Train Test Split on the Data

```python
train_set, test_set = train_test_split(census, test_size=0.3)
```

```python
X_train, y_train, X_test, y_test = train_set.iloc[:,:-1], train_set.iloc[:, -1].astype('float64'), test_set.iloc[:,:-1], test_set.iloc[:, -1].astype('float64')
```

** Normalize values of num columns**

```python
census.describe()
```

```python
num_cols = [i for i in census.columns if census[i].dtype!='O']
```

```python
scaler = preprocessing.MinMaxScaler()
scaler.fit(X_train[num_cols])
X_train_prep = scaler.transform(X_train[num_cols])
X_test_prep = scaler.transform(X_test[num_cols])
```

```python
X_train[num_cols] = X_train_prep
X_test[num_cols] = X_test_prep
```

```python
X_train.head()
```

### Create the Feature Columns for tf.esitmator

** Take note of categorical vs continuous values! **

```python
cols = census.columns[:-1]
cols
```

```python
feat_columns = [tf.feature_column.numeric_column(i) for i in cols if census[i].dtype!='O']+[tf.feature_column.categorical_column_with_vocabulary_list(i, list(census[i].unique())) for i in cols if census[i].dtype=='O']
feat_columns
```

```python
'''age_bucket  = tf.feature_column.bucketized_column(feat_columns[1], boundaries=[20,30,40,50,60,70,80])
age_bucket'''
```

### Create Input Function

** Batch_size is up to you. But do make sure to shuffle!**

```python
input_fn  = tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train.astype('float64'),batch_size=100,num_epochs=200,shuffle=True)
```

#### Create your model with tf.estimator

**Create a LinearClassifier.(If you want to use a DNNClassifier, keep in mind you'll need to create embedded columns out of the cateogrical feature that use strings, check out the previous lecture on this for more info.)**

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_columns)
```

** Train your model on the data, for at least 5000 steps. **

```python
model.train(input_fn=input_fn)
```

### Evaluation

** Create a prediction input function. Remember to only supprt X_test data and keep shuffle=False. **

```python
evaluation_fn  = tf.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size=10,num_epochs=1,shuffle=False)
results=model.evaluate(evaluation_fn)
```

```python
results
```

** Use model.predict() and pass in your input function. This will produce a generator of predictions, which you can then transform into a list, with list() **

```python
predictions = model.predict(input_fn=evaluation_fn)
```

** Create a list of only the class_ids key values from the prediction list of dictionaries, these are the predictions you will use to compare against the real y_test values. **

```python
ps = []
for p in list(predictions):
    ps.append(int(p['class_ids']))
```

** Import classification_report from sklearn.metrics and then see if you can figure out how to use it to easily get a full report of your model's performance on the test data. **

```python
from sklearn.metrics import classification_report

print(classification_report(ps, y_test))
```
