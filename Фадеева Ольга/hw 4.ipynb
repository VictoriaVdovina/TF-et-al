```python
import tensorflow as tf
```

```python
from sklearn import model_selection
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
import pandas as pd
df = pd.read_csv('CHD.csv', sep=',')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
split_train, split_test = model_selection.train_test_split(df, test_size=0.3)
```

```python
scaler = MinMaxScaler()
```

```python
test_xs = split_test.loc[:, 'HouseAge':'Longitude']
train_xs = split_train.loc[:, 'HouseAge':'Longitude']
```

```python
test_y = split_test['medianHouseValue']
train_y = split_train['medianHouseValue']
```

```python
train_fit = scaler.fit(train_xs)
```

```python
trans_train = scaler.transform(test_xs)
trans_test = scaler.transform(train_xs)
```

```python
age = tf.feature_column.numeric_column('HouseAge')
rooms = tf.feature_column.numeric_column('AveRooms')
bedrooms = tf.feature_column.numeric_column('AveBedrms')
population = tf.feature_column.numeric_column('Population')
occup = tf.feature_column.numeric_column('AveOccup')
income = tf.feature_column.numeric_column('MedInc')
```

```python
features = [age, rooms, bedrooms, population, occup, income]
```

```python
input_f = tf.estimator.inputs.pandas_input_fn(x=train_xs, y=train_y, batch_size=200, num_epochs = 10, shuffle=True)
```

```python
estim = tf.estimator.DNNRegressor(feature_columns=features, hidden_units=[2,3,4,5,6,5,4,3,2])
```

```python
estim.train(input_fn=input_f, steps=1000)
```

```python
predict_f = tf.estimator.inputs.pandas_input_fn(x=test_xs, batch_size=10, num_epochs = 1, shuffle=True)
predictions = estim.predict(input_fn=predict_f)
```

```python
predictions= list(predictions)
```

```python
preds = [i['predictions'] for i in predictions]
```

```python
from sklearn import metrics

metrics.mean_squared_error(preds, test_y)**0.5
```
